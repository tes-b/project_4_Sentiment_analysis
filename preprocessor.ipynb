Zoom

```python
def create_df_from_zoom(path:str, date:str):
    # print(path)
    with open(path) as file:
        lines = file.readlines()

    lst_date = []
    lst_time = []
    lst_author = []
    lst_cont = []
    time = ""
    author = ""
    
    for line in lines:
        if line.startswith("\t"):
            lst_date.append(date)
            lst_time.append(time)
            lst_author.append(author)
            lst_cont.append(line)
            # print(line)
        else: 
            info = line.split(" ")
            author = info[2]
            time = info[0]

    df = pd.DataFrame({
        "Date":lst_date,
        "Time":lst_time,
        "Author":lst_author,
        "Content":lst_cont
        })
    # print(zoom.shape)
    # zoom.head()
    return df
```

```python
cwd = os.getcwd()
data_path = cwd + "/" + "data/Zoom"
children = os.listdir(data_path)
# print(children)
lst_zoom = []
for child in children:
    # print(child)
    date = child.split(" ")[0]
    # print(date)
    file = os.listdir(data_path + "/" + child)[0]
    # print(file)
    path = data_path + "/" + child + "/" + file

    # print(path)
    lst_zoom.append(create_df_from_zoom(path, date))

len(lst_zoom)
lst_zoom[0]
```

```python
df_merge = pd.concat(lst_zoom)
print(df_merge.shape)
df_merge.head()
```

```python
def preprocessing_zoom(df):
    df_copy = df.copy()
    # df_copy["Author"] = df_copy["Author"].str[:-5]
    # list_date = df_copy["Date"].str.split(" ")
    
    # df_copy["Date"] = list_date.str.get(0)
    df_copy["Date"] = pd.to_datetime(df_copy["Date"])
    # df_copy["Time"] = list_date.str.get(1)
    df_copy["Content"] = df_copy["Content"].str.replace("\t","")
    df_copy["Content"] = df_copy["Content"].str.replace("\n","")
    df_copy = df_copy[["Date","Time","Author","Content"]]
    df_copy.sort_values(by=["Date","Time"], inplace=True)
    df_copy = df_copy.reset_index(drop=True)
    return df_copy
```

```python
df_zoom = preprocessing_zoom(df_merge)
print(df_zoom.shape)
df_zoom.head()
```
